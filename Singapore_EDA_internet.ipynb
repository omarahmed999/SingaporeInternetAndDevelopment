```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
import wbgapi as wb
import pandas as pd

# indicator code → friendly name
indicators = {
    "NY.GDP.MKTP.KD.ZG": "GDP_growth_annual_pct",
    "NY.GDP.PCAP.KD": "GDP_per_capita_const2015_USD",
    "BX.KLT.DINV.WD.GD.ZS": "FDI_net_inflows_pct_GDP",
    "SP.DYN.LE00.IN": "Life_expectancy_years",
    "SH.DYN.MORT": "Under5_mort_per1000",
    "SE.TER.ENRR": "School_enrollment_tertiary_pct",
    "SE.ADT.LITR.ZS": "Adult_literacy_pct",
    "SI.POV.GINI": "Gini_index",
    "GB.XPD.RSDV.GD.ZS": "R_and_D_pct_GDP",
    "TX.VAL.TECH.MF.ZS": "Hightech_exports_pct_manuf",
    "IT.NET.USER.ZS": "Internet_users_pct",
    "IT.CEL.SETS.P2": "Mobile_subs_per100"
}

# fetch by indicator codes; pass the dict directly
df = wb.data.DataFrame(indicators.keys(), economy='SGP', time=range(1960, 2025))

# rename columns using your friendly names
df.rename(columns=indicators, inplace=True)
df = df.reset_index().rename(columns={'economy': 'Country', 'time': 'Year'})

print(df.head())
```

```python
# Transpose
df_t = df.T

# Make the first row (which holds indicator codes) the header
df_t.columns = df_t.iloc[0]   # Use row 0 as header
df_t = df_t[1:]               # Drop the first row (already used as header)

# Now rename columns with meaningful names
df_t = df_t.rename(columns={
    "BX.KLT.DINV.WD.GD.ZS": "FDI_net_inflows_pct_GDP",
    "GB.XPD.RSDV.GD.ZS": "R_and_D_pct_GDP",
    "IT.CEL.SETS.P2": "Mobile_subs_per100",
    "IT.NET.USER.ZS": "Internet_users_pct",
    "NY.GDP.MKTP.KD.ZG": "GDP_growth_annual_pct",
    "NY.GDP.PCAP.KD": "GDP_per_capita_const2015_USD",
    "SE.ADT.LITR.ZS": "Adult_literacy_pct",
    "SE.TER.ENRR": "School_enrollment_tertiary_pct",
    "SH.DYN.MORT": "Under5_mort_per1000",
    "SI.POV.GINI": "Gini_index",
    "SP.DYN.LE00.IN": "Life_expectancy_years",
    "TX.VAL.TECH.MF.ZS": "Hightech_exports_pct_manuf"
})

# Reset index so "YR1960", "YR1961", ... become a column
df_t = df_t.reset_index().rename(columns={'index': 'Year'})

# Check the structure
df_t.head()
```

```python
df_t.info()
```

```python
df_dropped = df_t.drop('Gini_index', axis=1)
```

```python
df_dropped.info()
```

```python
df_dropped.head()
```

```python
df_dropped = df_t.drop(['School_enrollment_tertiary_pct','Hightech_exports_pct_manuf','Adult_literacy_pct','Gini_index'],axis=1)
```

```python
df_dropped.info()
```

```python
df_dropped.isnull().sum()
```

```python
df_dropped['Year'] = range(1960,2025)
df_dropped.head()
```

```python
cols_to_convert = df_dropped.columns[1:]
cols_to_convert
```

```python
df_dropped[cols_to_convert] = df_dropped[cols_to_convert].apply(
    lambda x: pd.to_numeric(x, errors='coerce'))
```

```python
df_dropped.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Compute correlation matrix (ignores NaN automatically)
corr = df_dropped.corr(numeric_only=True)

# Set up the figure
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0, fmt=".2f")
plt.title("Correlation Heatmap of Socio-Economic Indicators (Singapore)")
plt.savefig("heatmap_SNG")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=df_dropped, x="Year", y="Internet_users_pct", marker="o")
plt.title("Growth of Internet Users Over Time")
plt.ylabel("% of Internet Users")
plt.savefig("internetvstime")
plt.show()
```

```python
sns.scatterplot(data=df_dropped, x="Internet_users_pct", y="GDP_per_capita_const2015_USD")
plt.title("Internet Usage vs GDP per Capita")
plt.savefig("internetvsGDPperCapita")
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled = df_dropped.copy()
scaled[df_dropped.columns[1:]] = scaler.fit_transform(df_dropped[df_dropped.columns[1:]])
sns.lineplot(data=scaled, x="Year", y="Internet_users_pct", label="Internet")
sns.lineplot(data=scaled, x="Year", y="GDP_per_capita_const2015_USD", label="GDP per Capita")
sns.lineplot(data=scaled, x="Year", y="Life_expectancy_years", label="Life Expectancy")
plt.title("Normalized Growth Over Time")
plt.legend()
plt.savefig("NOrmaized")
plt.show()
```
